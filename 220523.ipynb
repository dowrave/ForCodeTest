```python
lst = "20 10 35 30 7"
num = list(map(int, lst.split(" ")))
N = len(num)

maxi = num[0]
mini = num[N-1]

for i in range(N):
    if num[i] > maxi:
        maxi = num[i]
    elif num[N-i-1] < mini:
        mini = num[N-i-1]

print(mini, maxi)
```

# Map 함수에 파라미터가 2개면 써지나?

- function에는 당연히 lambda도 올 수 있음
- 뒤에 오는 iterable에는 int나 float 값이 올 수 없음 -> 함수 안에 미리 포함시켜야 할 듯
- 한편 iterable한 복수의 argument는 그냥 ,로 나열해줄 수 있음

```python
Max = 75
scores = [75, 25, 30]

list(map(lambda scores : scores * Max / 100, scores))
```

### 백준 4344
1. 슬라이싱 범위 신경쓰기
2. 소수점 셋째짜리까지 나타내는 방법
    - round를 쓰면 40.000%가 아니라 40.0% 언더로 표시가 되지 않음
    - 이 때 쓰는 게 `포맷팅`임 : `{:.3f}.format(~~~)` - : 소숫점 셋째짜리"까지" 표시

```python
import sys

sys.stdin = open("test.txt", "r")
inputs = sys.stdin.readlines()

C = int(inputs[0]) # 테스트 케이스의 수
total_list = [0] * C

case = inputs[1:C + 1] # 슬라이싱 꼭 신경쓰기 : 실수가 잦음
# print(case)
for idx, lst in enumerate(case):
    
    lst = list(map(int, lst.split(" ")))
    students = lst[0]
    scores = lst[1:]
    sum = 0
    above_avg = 0
            
    for i in scores:
        sum += i
    avg = sum / students

    
    for j in scores:

        if j > avg:
            above_avg += 1
    above_avg_rate = above_avg / students
    total_list[idx] = above_avg_rate * 100

# 소수점 세번째자리까지 온전하게 출력.. 귀찮쓰
for score in total_list:
#     print(format(score, ".3f"))
    print("{:.3f}%".format(score))
```

```python
# 일단 행렬의 연산부터 보자

d_lst = []
num_lst = [i + 1 for i in range(10000)]

for i in num_lst:
    next_num = 0
    next_num = next_num + i
    while True:
        sub_num = i % 10
        next_num += sub_num
        i = i // 10
        if i == 0:
            break
    if next_num not in d_lst:
        d_lst.append(next_num)

for j in d_lst:
    
    if j in num_lst and j <= 10000:
        num_lst.remove(j)

for i in num_lst:
    print(i)
```

- 생각해야 할 점 : 전체 리스트에서 제거하려면 결국 한번씩 다 연산 해봐야 함

```python
# 각 자리수 추출
# N = 192
# div = 10
# a = N % 10
# b = (N // div % 10)
# c = (N // div**2 % 10)

# 위를 반복문으로 돌리자
N = 210
hansu_count = 0
if 1 <= N & N <= 99:
    hansu_count = N
elif 100 <= N & N <= 1000:
    hansu_count = 99
    for i in range(100, N+1):
        lst = []
        sub_N = i
        
        while True:
            num = sub_N % 10
            lst.append(num)
            sub_N = sub_N // 10
            if sub_N == 0:
                break
            
        lst = list(reversed(lst))
        
        
        if (lst[0] - lst[1]) == (lst[1] - lst[2]):
            hansu_count += 1
        # 1000까지니까 위 알고리즘이 먹힘 : 1000은 해당 없으니까
print(hansu_count)
```
