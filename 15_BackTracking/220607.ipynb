#### 시간 길게 끌리는 게 싫어서 일단 정답 찾아봄
1. 2차원 배열로 품 : 0의 위치를 별도의 리스트를 만들어 표시함
2. 그 0의 가로줄, 세로줄, 3*3에 없는 수를 넣으면 된다. - 즉 조건식을 3개 써야겠죠?


```python
import sys
# sys.stdin = open('test.txt', 'r')

lst = []
blanks = []
compare = [i for i in range(1, 10)]

for i in range(9):
    lst.append(list(map(int, sys.stdin.readline().rstrip('\n').split(' '))))

for i in range(9):
    for j in range(9):
        if lst[i][j] == 0:
            blanks.append([i, j])

print(blanks)
```

```python
# 조건식 만들자
# 자꾸 조건문 말고 == 넣는 습관좀 고치자
def horizontal_one(i, j):
    
    not_in = list(set(compare) - set(lst[i][:]))

    if len(not_in) == 1:
        lst[i][j] = not_in[0]
        print('horizontal : ' , lst[i][j])
        return True

def vertical_one(i, j):
    vertical_line = [m[j] for m in lst]
    not_in = list(set(compare) - set(vertical_line))
    print("Vertical Not In : ", not_in)
    if len(not_in) == 1:
        lst[i][j] = not_in[0]
        print('vertical : ' , lst[i][j])
        return True


def sqaure_one(i, j):
    
    # 2차원 리스트는 슬라이싱해도 2차원이다.. 그래서 3*3 행렬을 만들려면 구성을 이렇게 해준다.
#     x_boundary_min = i // 3 * 3 
#     x_boundary_max = i // 3 * 3 + 3 # 슬라이싱 할 것이므로 + 3까지 해줌
    
    comp_lst = [i[j // 3 * 3 : j // 3 * 3 + 3] for i in lst[i // 3 * 3 : i // 3 * 3 + 3]]

    not_in = list(set(compare) - set(list(sum(comp_lst, [])))) # 얘만 2차원이라 unhashable이 뜨는 것 같다.
    
    if len(not_in) == 1:
        lst[i][j] = not_in[0]
        print('square : ' , lst[i][j])
        
        
def condition_check(i, j):
    print(f"Case ({i}, {j})")
    horizontal_one(i, j)
    vertical_one(i, j)
    sqaure_one(i, j)
    print(" ")
```

```python
print('before : ')
for i in lst:
    print(i)

for blank in blanks:
    condition_check(blank[0], blank[1])
    
print('after : ')
for i in lst:
    print(i)
```

- 문제 풀면서 걸리적거렸던 지점
1. 어떤 값을 뽑아내고 재귀식을 원위치시킬 때, `value`값도 그 앞으로 되돌려야 함. 근데 global 선언하면 값을 뽑고 되돌리기가 어려움. 
    - 즉 어떤 Value를 썼는데, 그 사용한 Value를 앞의 재귀식으로 되돌리는 지점에서 막혔다는 뜻

```python
import sys
sys.stdin = open('test.txt', 'r')
input = sys.stdin.readline

N = int(input())
a_lst = list(map(int, input().rstrip('\n').split(' ')))
op = list(map(int, input().rstrip().split())) # 연산자 수

maximum = -1e9
minimum = 1e9


# 아예 이런 식으로 어떤 층에서 사용한 연산과 누적값, 남은 연산자의 수를 전부 파라미터로 때려넣을 수도 있다.
# depth는 결국 인덱스처럼 쓸 수 있는 값임
def dfs(depth, total, plus, minus, multiply, divide):
    global maximum, minimum
    if depth == N:
        maximum = max(total, maximum)
        minimum = min(total, minimum)
        return
    
    # elif문을 타지 않음 : plus 값이 남은 상황에서 다른 연산자가 먼저 올 수도 있기 때문에
    if plus:
        dfs(depth + 1, total + a_lst[depth], plus - 1, minus, multiply, divide)
    if minus:
        dfs(depth + 1, total - a_lst[depth], plus, minus - 1, multiply, divide)
    if multiply:
        dfs(depth + 1, total * a_lst[depth], plus, minus, multiply - 1, divide)
    if divide:
        dfs(depth + 1, int(total / a_lst[depth]), plus, minus, multiply, divide - 1)
        
# 첫 값은 어차피 연산이 없이 들어가므로, total은 0번째 인덱스부터 시작하면 됨
dfs(1, a_lst[0], op[0], op[1], op[2], op[3])
print(maximum)
print(minimum)
```
